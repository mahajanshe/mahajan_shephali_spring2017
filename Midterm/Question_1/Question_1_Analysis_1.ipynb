```python
#importing libraries
import glob
import re
import os
import email
import csv
from email.parser import Parser
from validate_email import validate_email
from nltk.corpus import stopwords
#Path to fetch files
path = glob.glob(r'C:\Users\Shephali Mahajan\Desktop\Python_Midterm\Enron_Data\Enron_Dataset\maildir\*')
```

```python
#Getting the mailIds
def from_mail_list(mails):
    for x in mails:
        if x[-1].isdigit():
            with open(x) as data_file:
                raw_file = data_file.read()
                email = Parser().parsestr(raw_file)
                email_list.append(re.sub("<.*?>","",email['From']))
    return email_list
```

```python
#Finding the frequency of words
def freq_list(mails):
    freq = {}
    for y in mails:
        if y not in freq:
            freq[y] = 1
        else:
            freq[y] += 1
    return freq
```

```python
#Getting the list of all emails
email_list = []
for folder_list in path:
    all_folders = glob.glob(os.path.join(folder_list, '*sent*'))
    for all_sent_folders in all_folders:
        all_mails = glob.glob(os.path.join(all_sent_folders, '*'))
        list_of_email = from_mail_list(all_mails)
        
    #                 emp_mail_Id = {} 
# (re.sub("<.*?>","",email['X-From']))
```

```python
#Cleaning emails from other domain
clean_emails = []
for x in list_of_email:
    if "@" in x:
        clean_emails.append(x)
```

```python
#Getting freqency of mail id's
enron_freq = freq_list(clean_emails)

#Sorting the emails with count in descending order
sortedlist = sorted(enron_freq.items(), key = lambda x:x[1], reverse = True)
sortedlist
```

```python
with open('Enron_mail_freq.csv', 'w', newline="") as csvfile:
    csv_output = csv.writer(csvfile)
    csv_output.writerow(['Mail Id', 'Freq'])
    for row in sortedlist:
        csv_output.writerow([row[0], row[1]])
        csvfile.close
```
