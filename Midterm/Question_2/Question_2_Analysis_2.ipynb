#Check the topics that appeared most frequently in headlines of articles published in 2014

```python
import glob   # imported glob
import os   # imported os
import csv  # imported csv
import json
from nltk.corpus import stopwords
files = glob.glob(r'C:\Users\Shephali Mahajan\Desktop\Python_Midterm\Question_2_files\U.S\*')  # storing readeable files in a variable
```

```python
abc = []
for p in files:     #for loop to read every file of the corpus
    with open(p, 'r') as f:   #open files
        json_file = json.load(f)
        json_keys = json_file.keys()
        for x in json_keys:
            if x == 'headline':
                y = json_file['headline']
                z = y.keys()
                for a in z:
                    if a == 'print_headline':
                        headline = json_file['headline']['print_headline'].split()
                        for word in headline:
                            abc.append(word)                                            
```

```python
#Lowercase the words in the list
abc = [x.lower() for x in abc]
```

```python
# Filtering out the stopwords
abc_filtered_stopwords = [word for word in abc if (word.lower()) not in stopwords.words('english')]
filtered_words = []
for a in abc_filtered_stopwords:
    d = a.replace('.','')
    filtered_words.append(d)
```

```python
#Counting the filtered words
from collections import Counter
counts = Counter(filtered_words)
```

```python
#Creating a list to save the data
dict_count= dict(counts)
dict_list = []
temp = []
for key, value in dict_count.items():
    temp = [key,value]
    dict_list.append(temp)
```

```python
#Most frequent topics on which the articles were published in 2014
for k,v in sorted(dict_count.items(), key =lambda x:x[1:] , reverse = True):
    x=  print(k, v) 
```
