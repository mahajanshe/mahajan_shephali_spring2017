-- Find out how many words and how many unique words in file 'austen-persuasion.txt'.

```python
from nltk.corpus import gutenberg
gutenberg.fileids()#show files
```

```python
ausper = gutenberg.words('austen-persuasion.txt')
print('Number of words in the file austen-persuasion.txt is',len(ausper)) #prints length of words in file 'austen-persuasion.txt'
```

```python
lower_case = [word.lower() for word in ausper]  #lowercase all words
len(set(lower_case)) #number of distinct words
```

```python
-- Calculate the frequency for each words and show the top 10 most frequent words and their frequency.
```

```python
count = {} # declared an empty dictionary
```

```python
for x in lower_case:   # for loop to count
    if x in count:
        count[x] +=1
    else:
        count[x] = 1
```

```python
count
```

```python
import operator    #sorting the key on basis of values in ascending order
sorted_x = sorted(count.items(), key=operator.itemgetter(1))
sorted_x
```

```python
top_10 = sorted_x[-10:] #sorted top 10 words
top_10
```
