```python
import nltk #importing nltk
```

```python
from nltk.corpus import stopwords #importing
#stopwords.words('english')
```

```python
from nltk.corpus import gutenberg #importing gutenerg from nltk corpus
```

```python
guten_files=gutenberg.fileids() # listdown all files in gutenberg
guten_files
```

```python
shakespeare_wordlist = [] #creating empty list
for file in guten_files:
    if "shakespeare" in file: # file belonging to Shakespeare's filtered out.
        word1= gutenberg.words(file)
        shakespeare_wordlist.extend(word1)   # turn to lower case
shakespeare_wordlist = [x.lower() for x in shakespeare_wordlist] #turn in lower case
#shakespeare_wordlist
#len(shakespeare_wordlist)
```

```python
Shake_filtered_stopwords = [word for word in shakespeare_wordlist if (word.lower()) not in stopwords.words('english')] # filtering out stopwords
#len(Shake_filtered_stopwords)
```

```python
import re  #import regular expression
nonPunct = re.compile('.*[A-Za-z].*')  # filter out punctutations, only characters present
shake_filtered_punct = [w for w in Shake_filtered_stopwords if nonPunct.match(w)]

Shake_final_counts1={}  #creating new dictionary
for b in shake_filtered_punct:
    if b not in Shake_final_counts1: 
        Shake_final_counts1[b] = 1 # if word appears first time, it will be added to the dictionary
    else:
        Shake_final_counts1[b] += 1 # if word reappears again it will be added to the same dictionary 
Shake_final_counts1
shake_filtered_punct
```

```python
from nltk import FreqDist #importing frequency distribution
```

```python
shakespeare=FreqDist(Shake_final_counts1)
```

```python
shakespeare.most_common(5)
```

```python
from nltk.corpus import webtext #importing webtext from nltk corpus
```

```python
web_file = webtext.fileids()
```

```python
webtext_wordlist = [] # creating empty list
for file1 in web_file:  # for loop
        word1= webtext.words(file1) 
        webtext_wordlist.extend(word1)
webtext_wordlist = [x.lower() for x in webtext_wordlist] #turn in lower case
#webtext_wordlist[:20]
```

```python
web_filtered_stopwords = [word for word in webtext_wordlist if (word.lower()) not in stopwords.words('english')]
#len(web_filtered_stopwords)
```

```python
nonPunct1 = re.compile('.*[A-Za-z].*')  # filter out punctutations and only contain letters
web_filtered_punct = [w for w in web_filtered_stopwords if nonPunct1.match(w)]

web_final_counts1={} #new dict is created
for a in web_filtered_punct: 
    if a not in web_final_counts1: 
        web_final_counts1[a] = 1 # if word appears first time, it will get added to the dictionary
    else:
        web_final_counts1[a] += 1 # if word reappears again it will get added to the same dictionary  
web_final_counts1
```

```python
web=FreqDist(web_final_counts1)
```

```python
web.most_common(5)
```

```python
shake_unusual_words={x:Shake_final_counts1[x] for x in Shake_final_counts1 if x not in web_final_counts1} #remove all entries from former if found in webtext dict 

shake_unusual_words
```

```python
web11=FreqDist(shake_unusual_words)
```

```python
ff= web11.most_common(5)
```

```python
# Find out what are the top 5 words that Shakespeare used but we do not use in currently.

print("Top 5 unusual words of Shakespears that are not used today:")
print("Word\tNumber of times appeared")

for TopFivewords in ff[0:5]: # Call top five words  
    print("{}\t{}".format(*TopFivewords))
```
