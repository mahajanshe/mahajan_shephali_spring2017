```python
# Importing all needed libraries
import glob, os, csv, datetime, email, string, re
import csv
import matplotlib.pyplot as plt
from email.parser import Parser
from nltk.tokenize import word_tokenize
```

```python
# Folder path
path = 'C:/Users/knave/Desktop/Python Assignments/Midterm/enron_mail_20150507/maildir/lay-k'
```

```python
#Creating empty list
words= []
```

```python
#Reading all inbox mails and storing in list
for root, dirs, files in os.walk(path, topdown=False):
    for name in files:
        with open(os.path.join(root, name)) as words:
            string = words.read()
            mail = Parser().parsestr(string)
            message = email.message_from_string(string)
            words= word_tokenize(string)
```

```python
# Removing punctuations and lowercase all words
x= [re.sub(r'[^A-Za-z0-9]+', '', a) for a in words]
lower = [a.lower() for a in x]
```

```python
# creating checklist of suspected words
checklist = ['Business', 'Program', 'enron','fraud', 'event', 'loss', 'profit','solutions', 'bankrupt']
```

```python
# matching the suspected words from the list and storing it in a new list
matches = [c for c in checklist if c in lower]
matches = []
for c in checklist:
    if c in lower:
        matches.append(c)
```

```python
# creating a dictionary to store the words and their freq used in conversation with Ken Lay
d= {}
for item in lower:
    if item in d:
        d[item] = d.get(item)+1
    else:
        d[item] = 1
dictlist = []
for key, value in d.items():
    temp = [key,value]
    dictlist.append(temp)        
sorting = sorted(d.items(), key = lambda x:x[1], reverse = True)
```

```python
# Storing the result in csv file
with open('Words_Freq.csv', 'w', newline="") as file:
    csv_output = csv.writer(file)
    csv_output.writerow(['Words', 'Frequency'])
    for row in sorting:
        csv_output.writerow([row[0], row[1]])
        file.close
```
